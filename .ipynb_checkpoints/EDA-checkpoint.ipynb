## Wind Direction

```python
def draw_heatmap(*args, **kwargs):
#     print(kwargs)
#     data = kwargs.pop('data')
#     print(data.head())
    try:
        data = kwargs.pop('data')
        d = data.reset_index().pivot_table(index=args[1], columns=args[0], values=args[2])
        sns.heatmap(d, **kwargs)
    except Exception as e:
        print(e)

g = sns.FacetGrid(train, col='meter',height=10, col_wrap=2,sharex=False, dropna=False)
g.map_dataframe(draw_heatmap,'site_id', 'wind_direction', 'meter_reading')
```

```python

def createWordcloud(*args, **kwargs):
    from wordcloud import WordCloud
    try:
        data = kwargs.pop('data')
        text = ' '.join([re.sub('(/| )','_',x) for x in data[args[0]].to_list()])
        wc = WordCloud(max_words=3).generate(text)
        plt.imshow(wc, interpolation='bilinear')
#         plt.axis("off")
    except Exception as e:
        print(e)
```
